```python
# import netcdf
netCDF_dir = r"C:\Users\Moos-Hyde\Documents\Hyde-Platform\data\netCDF"
netCDF_name = "irrigated_rice.nc"
nc = xr.open_dataset(os.path.join(netCDF_dir, netCDF_name))
nc
```

```python
# import a country
list(pycountry.countries)[0]
downloader = GADMDownloader(version="4.0")
ad_level = 0
country_shp = downloader.get_shape_data_by_country(list(pycountry.countries)[0], ad_level)
country_mask = regionmask.Regions(country_shp.geometry)
country_mask.mask(nc['lon'], nc['lat'])
country = nc.where(country_mask == 0)
```
